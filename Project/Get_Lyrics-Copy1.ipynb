```python
import pandas as pd
import pip
import urllib
import requests
from bs4 import BeautifulSoup as bs
```

```python
def scrape_lyrics(url):

    response = requests.get(url)
    html = response.text

    soup = bs(html, 'html.parser')

    lyrics = soup.find(name="lyrics")
 

    lyrics_text = ", ".join([lyric for lyric in lyrics.stripped_strings])
    return lyrics_text
```
